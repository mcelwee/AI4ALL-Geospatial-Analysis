# Exploratory Data Analysis (EDA) Notebook

# Getting Data:

```python
# Importing the libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
import pandas as pd
%matplotlib inline
```

```python
#Import Dataset
dataset = pd.read_csv('~/library/Iris.csv')
dataset.head()
```

# Summary Of the Dataset

```python
# shape
print(dataset.shape)
```

```python
# more info on the data
print(dataset.info())
```

```python
# descriptions
print(dataset.describe())
```

```python
# class distribution
print(dataset.groupby('iris').size())
```

# Visualizations:

```python
# box and whisker plots
dataset.plot(kind='box', sharex=False, sharey=False)
```

```python
# box and whisker plots
dataset.plot(kind='box', sharex=True, sharey=True)
```

```python
# histograms
dataset.hist(edgecolor='black', linewidth=1.2)
```

```python
# boxplot on each feature split out by species
dataset.boxplot(by="iris",figsize=(10,10))
```

```python
# violinplots on petal-length for each species
sns.violinplot(data=dataset,x="iris", y="petal length")
```

```python
from pandas.tools.plotting import scatter_matrix
# scatter plot matrix
scatter_matrix(dataset,figsize=(10,10))
plt.show()
```

```python
# Using seaborn pairplot to see the bivariate relation between each pair of features
sns.pairplot(dataset, hue="iris")
```

From the plot, we can see that the species setosa is separataed from the other two across all feature combinations
